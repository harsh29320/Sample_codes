```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('C:/Users/harshvardhan/Downloads/2017-18_AdvancedStats_Salary.csv')
y = data['WS']
BLK = data['BLK%']
```

```python
#Harsh code: import  seaborn
import seaborn as sns
```

```python
#Harsh code: make categories
def BLK_categories(BLK):
    
    if BLK < 2:
        return 1
    elif BLK > 2 and BLK < 4:
        return 2
    elif BLK > 4 and BLK < 6:
        return 3
    else:
        return 4
    
```

```python
data['BLK_categories'] = data['BLK%'].apply(BLK_categories)
```

```python
fig, ax = plt.subplots()
plt.scatter(BLK, y, color='black')
ax.set_xlabel('Block Percentage')
ax.set_ylabel('Win Shares')
```

```python
sns.lmplot(x='BLK%', y='WS', hue='BLK_categories', data=data)
ax.set_xlabel('Block Percentage')
ax.set_ylabel('Win Shares')
```
